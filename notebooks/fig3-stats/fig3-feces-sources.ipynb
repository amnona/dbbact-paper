```python
import calour as ca
```

```python
ca.set_log_level(11)
```

```python
dat=ca.read_amplicon('./all.feces-source.biom',
                     './map.feces-source.txt',normalize=10000,min_reads=1000)
```

```python
datc=dat.cluster_features(10)
```

```python
%matplotlib inline
```

```python
db=ca.database._get_database_class('dbbact')
```

```python
# datc=datc.add_terms_to_features('dbbact',max_id=3925)
```

# the dbbact version to use

```python
# max_id = 3925 # version 10-20
max_id = 6237 # version 05-21
```

# wordclouds

```python
tt=datc.filter_samples('host','cat')
tt=tt.filter_prevalence(0.3)
f=db.draw_wordcloud(tt,width=1600,height=1200,max_id=max_id,threshold=0.1,relative_scaling=0)
```

```python
f.savefig('sources-cat.pdf',dpi=600)
```
